## In this notebook I will
1. Pull data from the [Deportation Data Project]("https://deportationdata.org/data/ice.html") 
2. Extract ICE Arrests for 2022, 2023, 2024, 2025, 2026 
3. Clean the data for each year. 

```python
## import necessary libraries
import pandas as pd
```

### EXTRACTION1: Pulling 2022 data

```python
## let's pull data that i have downloaded. starting with the first file
df1 = pd.read_excel('/Users/sajinashrestha/Documents/dataProjects/WAMC/ICE Arrests/CLEANED-ICEarrests-Albany/RawData/2026-ICLI-00005_Arrests_FY23_20260311_Redacted.xlsx')
df1
```

```python
#### As you can see: we don't need the first couple rows, it has unneeded rows. Let's remove the rows
df1 = df1.iloc[5:]
df1
```

```python
## the first row is actually the header we want. Let's drop the header and put the actual header we want there
## create a new df, just to be safe
df1.columns = df1.iloc[0]
df1_titled = df1[1:].reset_index(drop=True)
df1_titled
```

```python
## to be safe, let's turn the time, which is what we need, into datetime
## turn the column into datetime
df1_titled['Apprehension Date'] = pd.to_datetime(df1_titled['Apprehension Date'])
```

```python
## create a subset of data for only 2022
##first: pull anything before 2023
df_2022 = df1_titled[df1_titled['Apprehension Date'] < '2023-01-01']
```

```python
##now: pull anything after 2021 from the same subset. 
df_2022 = df_2022[df_2022['Apprehension Date'] > '2021-12-31']
```

```python
##This is your 2022 ICE arrests data
df_2022
```

```python
##Let us do Extraction 1: 2022 ICE Arrests data
df_2022.to_csv('FilteredData/2022ICEArrestsData.csv', index=False)
```
