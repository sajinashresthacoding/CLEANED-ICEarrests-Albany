## In this notebook
1. Pull and combine all 2025 arrests data
2. Analyze it for:

    a. arrests in year by state for 2025
    
    b. arrests in NY for 2025
    
    c. arrests in Capital Region for 2025

```python
## import necessary libraries
import pandas as pd
```

```python
df1 = pd.read_csv('/Users/sajinashrestha/Documents/dataProjects/WAMC/ICE Arrests/CLEANED-ICEarrests-Albany/FilteredData/2025ICEArrestsDataPT.1.csv')
df1
```

```python
df2 = pd.read_csv('/Users/sajinashrestha/Documents/dataProjects/WAMC/ICE Arrests/CLEANED-ICEarrests-Albany/FilteredData/2025ICEArrestsDataPT.2.csv')
df2
```

```python
df25 = pd.concat([df1, df2])
df25
```

```python
df25_state = df25["State"].value_counts()
df25_state
```

```python
df25_state.to_csv('FilteredData/2025DeportationByState.csv', index=True)
```

```python
df_NY = df25.query('State == "NEW YORK"') 
df_NY
```

```python
NYcounts = df_NY['Apprehension Site Landmark'].value_counts()
NYcounts
```

```python
NYcounts.to_csv('FilteredData/2025DeportationByNY.csv', index=True)
```

```python
df_NY['Apprehension Site Landmark'].unique()
```

```python
CapitalRegion= ['ALBANY COUNTY','RENSSELAER COUNTY', 'SARATOGA COUNTY', 'GREENE COUNTY', 'WARREN COUNTY', 'WARREN COUNTY JAIL', 'SCHENECTADY COUNTY JAIL', 'SARATOGA COUNTY JAIL', 'RENSSELAER COUNTY JAIL', 'WASHINGTON CORRECTIONAL FACILITY', 'WASHINGTON COUNTY JAIL']
```

```python
df_CR = df_NY[df_NY['Apprehension Site Landmark'].isin(CapitalRegion)]
df_CR
```

```python
df_CR["Apprehension Site Landmark"].value_counts()
```

```python
df_CR.to_csv('FilteredData/2025CapitalRegionICEArrests.csv', index = True)
```
