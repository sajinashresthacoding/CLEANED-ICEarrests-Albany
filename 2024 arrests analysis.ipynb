## In this notebook
1. Pull and combine all 2024 arrests data
2. Analyze it for:

    a. arrests in year by state for 2024
    
    b. arrests in NY for 2024
    
    c. arrests in Capital Region for 2024

```python
## import necessary libraries
import pandas as pd
```

```python
df1 = pd.read_csv('/Users/sajinashrestha/Documents/dataProjects/WAMC/ICE Arrests/CLEANED-ICEarrests-Albany/FilteredData/2024ICEArrestsDataPT.1.csv')
df1
```

```python
df2 = pd.read_csv('/Users/sajinashrestha/Documents/dataProjects/WAMC/ICE Arrests/CLEANED-ICEarrests-Albany/FilteredData/2024ICEArrestsDataPT.2.csv')
df2
```

```python
df24 = pd.concat([df1, df2])
df24
```

```python
df24_state = df24["State"].value_counts()
df24_state
```

```python
df24_state.to_csv('FilteredData/2024DeportationByState.csv', index=True)
```

```python
df_NY = df24.query('State == "NEW YORK"') 
df_NY
```

```python
NYcounts = df_NY['Apprehension Site Landmark'].value_counts()
NYcounts
```

```python
NYcounts.to_csv('FilteredData/2024DeportationByNY.csv', index=True)
```

```python
df_NY['Apprehension Site Landmark'].unique()
```

```python
CapitalRegion= ['ALBANY COUNTY','RENSSELAER COUNTY', 'SARATOGA COUNTY', 'GREENE COUNTY', 'WARREN COUNTY', 'WARREN COUNTY JAIL', 'SCHENECTADY COUNTY JAIL', 'SARATOGA COUNTY JAIL', 'RENSSELAER COUNTY JAIL', 'WASHINGTON CORRECTIONAL FACILITY', 'WASHINGTON COUNTY JAIL']
```

```python
df_CR = df_NY[df_NY['Apprehension Site Landmark'].isin(CapitalRegion)]
df_CR
```

```python
df_CR["Apprehension Site Landmark"].value_counts()
```
