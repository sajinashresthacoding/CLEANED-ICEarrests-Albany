```python
## import necessary libraries
import pandas as pd
```

```python
df = pd.read_csv('/Users/sajinashrestha/Documents/dataProjects/WAMC/ICE Arrests/CLEANED-ICEarrests-Albany/FilteredData/2026ICEArrestsData.csv')
df
```

```python
df26_state = df["State"].value_counts()
df26_state
```

```python
df_NY = df.query('State == "NEW YORK"') 
df_NY
```

```python
NYcounts = df_NY['Apprehension Site Landmark'].value_counts()
NYcounts
```

```python
NYcounts.to_csv('FilteredData/2026DeportationByNY.csv', index=True)
```
