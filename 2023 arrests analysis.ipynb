## In this notebook
1. Pull and combine all 2023 arrests data
2. Analyze it for:

    a. arrests in year by state for 2023
    
    b. arrests in NY for 2023
    
    c. arrests in Capital Region for 2023

```python
## import necessary libraries
import pandas as pd
```

```python
df1 = pd.read_csv('/Users/sajinashrestha/Documents/dataProjects/WAMC/ICE Arrests/CLEANED-ICEarrests-Albany/FilteredData/2023ICEArrestsDataPT.1.csv')
df1
```

```python
df2 = pd.read_csv('/Users/sajinashrestha/Documents/dataProjects/WAMC/ICE Arrests/CLEANED-ICEarrests-Albany/FilteredData/2023ICEArrestsDataPT.2.csv')
df2
```

```python
df23 = pd.concat([df1, df2])
df23
```

```python
df23_state = df23["State"].value_counts()
df23_state
```

```python
df23_state.to_csv('RawData/2023DeportationByState.csv', index=True)
```

```python
df_NY = df23.query('State == "NEW YORK"') 
df_NY
```

```python
NYcounts = df_NY['Apprehension Site Landmark'].value_counts()
NYcounts
```

```python
NYcounts.to_csv('FilteredData/2023DeportationByNY.csv', index=True)
```

```python
df_NY['Apprehension Site Landmark'].unique()
```

```python
CapitalRegion= ['ALBANY COUNTY','RENSSELAER COUNTY', 'SARATOGA COUNTY', 'GREENE COUNTY', 'WARREN COUNTY', 'WARREN COUNTY JAIL', 'SCHENECTADY COUNTY JAIL', 'SARATOGA COUNTY JAIL', 'RENSSELAER COUNTY JAIL', 'WASHINGTON CORRECTIONAL FACILITY', 'WASHINGTON COUNTY JAIL']
```

```python
df_CR = df_NY[df_NY['Apprehension Site Landmark'].isin(CapitalRegion)]
df_CR
```

```python
df_CR["Apprehension Site Landmark"].value_counts()
```
